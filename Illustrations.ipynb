```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
# c=192; 4 lrs; 5 lrs, 6 lrs; pp 25%; pp 75%;
AUCs = np.array([96.7, 95.5, 94.5, 94.5, 95.2, 94.2])
pAUCs = np.array([93.7, 91.8, 91.7, 92.4, 92.1, 88.4])
colors = ['darkseagreen', 'royalblue', 'darkorchid', 'orchid', 'cadetblue', 'deepskyblue']
```

```python
plt.rcParams.update({'font.size': 15})
figure, axis = plt.subplots(1, 2, figsize=(10, 5))
axis[0].bar(range(1, len(AUCs) + 1), AUCs - 93.3, color=colors, edgecolor='black')
axis[0].set_title("AUC")
axis[0].set_ylabel("приращение в %")
axis[0].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
axis[0].yaxis.set_major_locator(MaxNLocator(integer=True))
#axis[0].set_xlabel("номер успешной модификации")


axis[1].bar(range(1, len(AUCs) + 1), pAUCs - 87.0, color=colors, edgecolor='black')
axis[1].set_title("pAUC")
axis[1].set_ylabel("приращение в %")
axis[1].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
axis[1].yaxis.set_major_locator(MaxNLocator(integer=True))
#axis[1].set_xlabel("номер успешной модификации")

channel_patch = mpatches.Patch(color='darkseagreen', label='Количество каналов увеличено в 1.5 раза')
plus_one_patch = mpatches.Patch(color='royalblue', label='Количество слоев в сети увеличено на 1')
plus_two_patch = mpatches.Patch(color='darkorchid', label='Количество слоев в сети увеличено на 2')
plus_three_patch = mpatches.Patch(color='orchid', label='Количество слоев в сети увеличено на 3')
percent_decrease_patch = mpatches.Patch(color='cadetblue', label='Процент перекрытий снижен до 25%')
percent_increase_patch = mpatches.Patch(color='deepskyblue', label='Процент перекрытий увеличен до 75%')
axis[1].legend(handles=[channel_patch, plus_one_patch, plus_two_patch, plus_three_patch, percent_decrease_patch, percent_increase_patch],
               bbox_to_anchor=(1.05, 0.65))

plt.show()
```
